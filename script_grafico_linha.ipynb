```python
''' Programa em Python que gera um gráfico em linha com dados estáticos.
É necessário instalar e/ou importar a biblioteca matplotlib.pyplot.

Exemplo de instalação: pip install matplotlib

Autor: Daniel Caixeta - @dfcaixeta - 30.mar.24 '''

# Importação da biblioteca.
import matplotlib.pyplot as plt

# Dados de entrada (eixos x e y).
x = [10, 20, 30, 40, 50]
y = [10, 20, 30, 40, 50]

# Define a cor da linha
plt.plot(x, y, c="green")

# Função usada para adicionar título principal e eixos (x e y) no gráfico.
plt.title("Gráfico de linha")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.grid(c='black', ls='--', lw=0.4)

# Esta função imprime/plota o gráfico.
plt.show()
```
