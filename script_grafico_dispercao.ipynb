```python
''' Programa em Python que gera um gráfico de dispersão em dados estáticos.
É necessário importar a biblioteca matplotlib.pyplot. Em caso de não instalada, instale
conforme o exemplo: pip install matplotlib.pyplot

autor: Daniel Caixeta - @dfcaixeta - 30.mar.2024

'''

# Importação da biblioteca.
import matplotlib.pyplot as plt

# Dados de entrada (eixos x e y).
x = [10, 20, 30, 40, 50, 60, 70, 80, 90]
y = [5, 20, 40, 10, 30, 70, 90, 70, 80]

# Definição das cores dos plots.
cores = ["red", "red", "blue", "red", "blue", "green", 
         "green", "orange", "orange"]

''' Observação importante! o número de cores declaradas tem que ser igual ao dos pontos
no gráfico. No exemplo acima, x = 9 pontos, y = 9 pontos, cores = 9) '''

# Esta função define as cores dos pontos no gráfico usando uma matriz definida ou não.
plt.scatter(x, y, c=cores) # Plota os pontos com as cores definidas na matriz. Usa-se o parâmetro c (colors).
# Por exemplo: plt.scatter(x, y, c="green") # Plota os pontos com a cor verde "green".

# Função usada para adicionar um título no gráfico.
plt.title("Gráfico de dispersão")

# Função que define um rótulo para o eixo x.
plt.xlabel("Eixo x")

# função que define um rótulo para o eixo y.
plt.ylabel("Eixo y")

# Esta função imprime/plota o gráfico
plt.show()
```
